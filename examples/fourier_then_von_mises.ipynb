# First let's define the unitary matrix/circuit whose eigenphase we want to infer

We'll use the second example from [Moore et al](https://www.chem.purdue.edu/kais/docs/publications/2021/Statistical%20approach%20to%20qpe.pdf) which takes the form $U = P(\pi /4) \otimes (H \cdot R_z(\pi/2) \cdot H)$ where
$$
P(\theta) =
\begin{pmatrix}
1 & 0 \\
0 & e^{i\theta}
\end{pmatrix}
,\qquad
R_z(\theta) =
\begin{pmatrix}
e^{-i\theta / 2} & 0 \\
0 & e^{i\theta / 2}
\end{pmatrix}
,\qquad
H = \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 & 1 \\
1 & -1
\end{pmatrix}
$$

```python
def P(theta):
    return jnp.array([[1, 0], [0, jnp.exp(1.j * theta)]])

def Rz(theta):
    return jnp.array([[jnp.exp(- 0.5j * theta), 0], [0, jnp.exp(0.5j * theta)]])

H = jnp.array([[1, 1], [1, -1]]) / jnp.sqrt(2)
```

```python
U = jnp.kron(P(jnp.pi / 4), H @ Rz(jnp.pi/2) @ H)
```

```python
U
```

We'll also get the exact eigenstates and eigenvalues (which we can do since this is a small scale example)

```python
true_eigvals, true_eigvecs = jnp.linalg.eig(U)
phi_ind = 0
phi_vec = true_eigvecs[:, phi_ind]
true_phi = jnp.arccos(true_eigvals[phi_ind].real)
print('True phase:', true_phi)
```

```python
# check true_phi satisfies eigenvector equation
jnp.allclose(U @ phi_vec, jnp.exp(1.j * true_phi) * phi_vec)
```

# Now we'll define a protocol for simulating measurements (without noise) from a phase estimation circuit

Although we'll assume rounds are run in parallel rather than on the same device (this does not change the likelihood but is easier to sample classically).

```python
N_U = U.shape[0]
n_qubits_U = int(jnp.log2(N_U))

big_H = jnp.kron(H, jnp.eye(N_U))
in_statevector = jnp.kron(jnp.array([1, 0]), phi_vec)

def experiment_probs(k: int, beta: float) -> jnp.ndarray:
    big_Rz = jnp.kron(Rz(beta), jnp.eye(N_U))
    U_k = jnp.linalg.matrix_power(U, k)
    controlled_U_k = jnp.block([[jnp.eye(N_U), jnp.zeros(U.shape)], [jnp.zeros(U.shape), U_k]])
    
    out_statevector = big_H @ controlled_U_k @ big_Rz @ big_H @ in_statevector
    measurement_probs = jnp.abs(out_statevector) ** 2
    measurement_probs_q1 = measurement_probs.reshape((2, ) * (n_qubits_U + 1)).sum(axis=tuple(range(1, n_qubits_U + 1)))
    return measurement_probs_q1


def run_experiment(k: int, beta: float, n_shots: int, random_key: random.PRNGKeyArray) -> jnp.ndarray:
    probs = experiment_probs(k, beta)
    return random.choice(random_key, a=jnp.arange(2), p=probs, shape=(n_shots,))
```

```python
# check experiment_probs gives correct likelihood
k_test = 4
beta_test = 0.43543
jnp.isclose(experiment_probs(k_test, beta_test)[0], 0.5 * (1 + jnp.cos(k_test * true_phi + beta_test)))
```

```python
# check samples give correct likelihood
shots = run_experiment(k_test, beta_test, 10000, random.PRNGKey(0))
jnp.isclose(1 - shots.mean(), 0.5 * (1 + jnp.cos(k_test * true_phi + beta_test)), atol=1e-2)
```

# Now we can run phase estimation

Follwing [O'Brien et al](https://iopscience.iop.org/article/10.1088/1367-2630/aafb8e/pdf), at each iteration we sample $\beta \sim [0, 2 \pi]$ and set $k = \lceil 1.25 / \sigma \rceil$ where $\sigma^2$ is the prior Holevo phase variance.

```python
def phase_estimation_iteration(prior_fourier_coeffs: jnp.ndarray, n_shots: int, random_key: random.PRNGKeyArray) -> jnp.ndarray:
    beta = random.uniform(random_key, minval=0, maxval=2 * jnp.pi)
    prior_holevo_phase_variance = phayes.fourier_holevo_variance(prior_fourier_coeffs)
    k = int(jnp.maximum(jnp.ceil(1.25 / jnp.sqrt(prior_holevo_phase_variance)), 1))

    shots = run_experiment(k, beta, n_shots, random_key)

    posterior_fourier_coeffs = phayes.fourier_update(prior_fourier_coeffs, shots, k, beta)
    return beta, k, posterior_fourier_coeffs
```

```python
n_shots_per_experiment = 10
num_experiments = 5
fourier_length = 200
```

```python
fcs = [jnp.zeros((2, fourier_length))]
ks = []
betas = []
```

```python
random_keys = random.split(random.PRNGKey(0), num_experiments)
```

```python
for i in range(num_experiments):
    beta, k, new_fc = phase_estimation_iteration(fcs[i], n_shots_per_experiment, random_keys[i])
    
    print(f"Experiment {i + 1} of {num_experiments}, k={k}, beta={beta:.2f}, coeffs used: {(new_fc != 0).sum() // 2}/{new_fc[0].size}")

    fcs.append(new_fc)
    ks.append(k)
    betas.append(beta)
```

```python
pdf_linsp = jnp.linspace(0, 2 * jnp.pi, 1000)

fig, axes = plt.subplots(num_experiments + 1, figsize=(5, 10))
for i in range(num_experiments + 1):
    axes[i].plot(pdf_linsp,phayes.fourier_pdf(pdf_linsp, fcs[i]), c='steelblue', label=f'p($\phi \mid m)$')
    axes[i].axvline(true_phi, color='green', label=r'True Phase $\phi$')
    axes[i].set_title(f"Experiment {i} of {num_experiments}")
    axes[i].set_xlim(0, 2 * jnp.pi)
    axes[i].set_ylim(0, axes[i].get_ylim()[1])
axes[-1].set_xlabel(r'$\phi$')
axes[-1].legend()
fig.tight_layout()
```

# Switch to von Mises representation

At this stage the posterior has become quite concentrated but we may still desire the phase to higher precision. We can check how many Fourier coefficients we have used so far

```python
print(f'Fourier coefficients used: {(new_fc != 0).sum() // 2}/{new_fc[0].size}')
```

Still no truncation, but with further experiments we would have to either increase the size of the array or truncate the series (which will lead to an invalid distribution/negative probabilities).

Instead we can convert the posterior to a von Mises approximation which only has two parameters and update this instead`

```python
von_mises_params = phayes.fourier_to_von_mises(new_fc)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].plot(pdf_linsp, phayes.fourier_pdf(pdf_linsp, new_fc), c='steelblue', label='Fourier')
axes[0].plot(pdf_linsp, phayes.von_mises_pdf(pdf_linsp, *von_mises_params), c='purple', label='von Mises')
axes[0].axvline(true_phi, color='green', label=r'True Phase $\phi$')
axes[0].set_xlabel(r'$\phi$')
axes[0].legend()

zoom_linsp = jnp.linspace(true_phi - 0.5, true_phi + 0.5, 1000)
axes[1].plot(zoom_linsp, phayes.fourier_pdf(zoom_linsp, new_fc), c='steelblue', label='Fourier')
axes[1].plot(zoom_linsp, phayes.von_mises_pdf(zoom_linsp, *von_mises_params), c='purple', label='von Mises')
axes[1].axvline(true_phi, color='green', label=r'True Phase $\phi$')
axes[1].set_xlabel(r'$\phi$ (zoomed in)')
axes[1].legend();
```

The approximation is not perfect due to the multimodality of the (true) Fourier posterior, however is suitable considering we now only need to store two parameters.

Let's run some more experiments

```python
def phase_estimation_iteration_vm(mu: float, kappa: float, n_shots: int, random_key: random.PRNGKeyArray) -> jnp.ndarray:
    beta = random.uniform(random_key, minval=0, maxval=2 * jnp.pi)
    prior_holevo_phase_variance = phayes.von_mises_holevo_variance(kappa)
    k = int(jnp.maximum(jnp.ceil(1.25 / jnp.sqrt(prior_holevo_phase_variance)), 1))

    shots = run_experiment(k, beta, n_shots, random_key)

    posterior_vm_params = phayes.von_mises_update(mu, kappa, shots, k, beta)
    return beta, k, posterior_vm_params
```

```python
num_vm_experiments = 5
vmps = [von_mises_params]
random_keys = random.split(random.PRNGKey(1), num_vm_experiments)
```

```python
for i in range(num_vm_experiments):

    beta, k, new_vmp = phase_estimation_iteration_vm(*vmps[i], n_shots_per_experiment, random_keys[i])
    
    print(f"von Mises experiment {i + 1} of {num_vm_experiments}, k={k}, beta={beta:.2f}")

    vmps.append(new_vmp)
    ks.append(k)
    betas.append(beta)
```

```python
fig, axes = plt.subplots(num_vm_experiments + 1, figsize=(5, 10))
for i in range(num_experiments + 1):
    axes[i].plot(zoom_linsp, phayes.von_mises_pdf(zoom_linsp, *vmps[i]), c='purple', label=f'p($\phi \mid m)$')
    axes[i].axvline(true_phi, color='green', label=r'True Phase $\phi$')
    axes[i].set_title(f"von Mises experiment {i} of {num_experiments}")
    axes[i].set_ylim(0, axes[i].get_ylim()[1])
axes[-1].set_xlabel(r'$\phi$')
axes[-1].legend()
fig.tight_layout()
```
